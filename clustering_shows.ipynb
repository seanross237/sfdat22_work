```python
just_shows1 = pd.read_csv('~/desktop/comon_baby2.csv')

just_shows1 = just_shows1.iloc[:10000,:16]


just_shows1.head()
```

```python
all_shows=[]

for x in range(1,2):
    eval('just_shows1.show' + str(x)+'.apply(lambda x: all_shows.append(x))')

    
#decided to only include shows that were someones top show at least once
total_shows = set(all_shows)
print len(total_shows)

show_dict_list=[] 

#create dictionary for each row in the data set, with 0's as values for each show in total shows
for row in range(len(just_shows1)): 
    show_dict_list.append({x:0 for x in total_shows})
print 'made dict'

#go through each tv's top shows and update that rows dictionary if the show matches. Shows are weighted to have more
#influence if it is their top show.
for row in range(len(just_shows1)):
    
    for show in show_dict_list[row]:
        for show_in_row in range(1,16):
            if just_shows1.iloc[row][show_in_row] == show:
                show_dict_list[row][show] = (1 * ((16-show_in_row)/3))
           
    if row % 20 == 0:
        print row
        #     if (just_shows1.iloc[row][7] == genre) or (data.iloc[row][8] == genre) or (data.iloc[row][9] == genre):
#            genre_dict_list[row][genre] = 1 
#             if row % 100 == 0:
#                 print row
len(total_shows)
```

```python
show_df = pd.DataFrame.from_records(show_dict_list)

df_for_cluster = pd.concat([just_shows1[['sdid']], show_df], axis=1)
```
